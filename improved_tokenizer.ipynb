```python
with open("Data/01 Harry Potter and the Sorcerers Stone.txt",'r',encoding="utf-8") as f:
    raw_text=f.read()
```

```python
import re
class TokenizerV2():
    def __init__(self,raw_text):
        clean_text=self._preprocess(raw_text)
        all_tokens=sorted(set(clean_text))
        all_tokens.extend(["<|endoftext|>", "<|unk|>"])
        vocab={token:index for index,token in enumerate(all_tokens)}
        self.str_to_ind=vocab
        self.ind_to_str={ind:token for token,ind in vocab.items()}

    def _preprocess(self,text):
        preprocessed=re.split(r'([,.:;?_!"()\']|--|\s)',text)
        result=[text.strip() for text in preprocessed if text.strip()]
        return [item.lower() for item in result]


    def encode(self,text):
        result=self._preprocess(text)
        result=[item if item in self.str_to_ind else "<|unk|>" for item in result]
        ids=[self.str_to_ind[item] for item in result]
        return ids
    
    def decode(self,ids):
        words=" ".join([self.ind_to_str[id] for id in ids])
        words = re.sub(r'\s+([,.?!"()\'])', r'\1', words)
        return words
        
```
