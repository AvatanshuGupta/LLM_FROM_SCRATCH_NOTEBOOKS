### Making token embedding matrix of the input tensor

```python
token_embedding=embedding_layer(input)
```

```python
token_embedding.shape
```

### Making positional embedding matrix

```python
context_length=4

positional_embedding_layer=torch.nn.Embedding(context_length,vec_dim)
```

```python
positional_embedding=positional_embedding_layer(torch.arange(context_length))
```

```python
positional_embedding
```

```python
positional_embedding.shape
```

### Adding token and positional embeddidng to get final input embedding

```python
final_input_embedding=token_embedding+positional_embedding
```
